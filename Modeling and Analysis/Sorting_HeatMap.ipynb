```python
import pandas as pd

# Step 1: Load the Excel file
df = pd.read_excel(r"C:\Users\incha\MTdata\Indian_Community\FPLM_Core_FVA_match.xlsx")  # Replace with your filename

# Step 2: Drop rows where Core Function is missing
df = df.dropna(subset=["Core Function"])

# Step 3: Split the 'Core Function' column on '|'
df["Core Function"] = df["Core Function"].str.split("|")

# Step 4: Explode the list into separate rows
df = df.explode("Core Function")

# Step 5: Strip whitespace around core function names
df["Core Function"] = df["Core Function"].str.strip()

# Step 6: Sort alphabetically by Core Function (without merging rows)
df_sorted = df.sort_values(by="Core Function").reset_index(drop=True)

# Step 7: Save to new Excel or CSV file
df_sorted.to_excel(r"C:\Users\incha\MTdata\Indian_Community\Sorted_FPLM_Core_FVA_match.xlsx", index=False)
# Or use df_sorted.to_csv("sorted_core_functions.csv", index=False)

print("Processing complete. Sorted file saved as 'sorted_core_functions.xlsx'.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Excel file
df = pd.read_excel(r"C:\Users\incha\MTdata\Indian_Community\Sorted_FPLM_Core_FVA_match.xlsx")  # Replace with your actual file

# Count the occurrences of each function
function_counts = df['Core Function'].value_counts().sort_values()

# Prepare data for heatmap
heatmap_data = pd.DataFrame(function_counts).T
heatmap_data.columns.name = 'Core Function'

# Set a larger figure size to accommodate long x-labels
plt.figure(figsize=(len(function_counts) * 0.8, 3.5))

# Create the heatmap
sns.set(font_scale=0.8)  # Adjust if labels are too large/small
ax = sns.heatmap(heatmap_data, cmap="YlGnBu", annot=True, fmt='d', cbar=True)

# Title and label formatting
plt.title("Frequency of Core Functions")
plt.xticks(rotation=45, ha='right')
plt.yticks([])  # Optional: remove the single y-tick

# Ensure layout is adjusted BEFORE saving
plt.tight_layout()

# Save high-res image exactly as shown in notebook
plt.savefig(r"C:\Users\incha\MTdata\Indian_Community\FPLM_Core_FVA_match_heatmap.png", dpi=300)

# Also display inline
plt.show()
```
