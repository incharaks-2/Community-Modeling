```python
from cobra.io import read_sbml_model
from pathlib import Path

# Path to your SBML model
model_path = Path(r"C:\Users\incha\Downloads\Abiotrophia_defectiva_ATCC_49176.xml") #Copy the address of the .xml as path and paste it.

# Check if the file exists
if model_path.exists():
    try:
        # Load the model
        model = read_sbml_model(str(model_path.resolve()))
        print("Model loaded successfully!")
    except Exception as e:
        print(f"Error loading SBML model: {e}")
else:
    print(f"File not found: {model_path}")
print(len(model.reactions))
print(len(model.metabolites))
print(len(model.genes))
model
```
